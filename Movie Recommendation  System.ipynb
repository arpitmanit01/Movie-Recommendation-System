```python
matrix=data.pivot_table(index='user id',columns='Title',values='rating')
```

```python
matrix
```

```python
similar_df=pd.DataFrame(matrix.corrwith(matrix['Star Wars (1977)']),columns=["Correlation"]).dropna()
```

```python
similar_df.reset_index()
suggest=pd.merge(left=similar_df.reset_index(),right=ratings.reset_index(),on='Title')
```

```python
suggest[suggest['Number of Rating']>100].sort_values('Correlation',ascending=False)
```
